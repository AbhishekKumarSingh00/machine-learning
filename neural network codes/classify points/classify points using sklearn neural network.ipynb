```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv('./Training Data/Logistic_X_Train.csv')
x_test=pd.read_csv('./Test Cases/Logistic_X_Test.csv').values
Y=pd.read_csv('./Training Data/Logistic_Y_Train.csv')
```

```python
X.info()
Y.info()
```

```python
print(X.head(7))
print(Y.head(7))
print(X.shape,Y.shape)
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
Y=Y.reshape((Y.shape[0],))
```

```python
Y.shape
```

```python
split=int(X.shape[0]*.8)
print(split)
```

```python
plt.scatter(X[:split,0],X[:split,1],c=Y[:split])
plt.show()
```

```python
plt.scatter(X[split:,0],X[split:,1],c=Y[split:])
plt.show()
```

```python
from keras import models
from keras.layers import Dense
```

```python
model=models.Sequential()
model.add(Dense(8,activation='relu',input_shape=(2,)))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(X,Y,batch_size=52,epochs=1000,validation_split=0.2)
```

```python
h=hist.history
```

```python
plt.plot(h['val_loss'],label="validation loss")
plt.plot(h['loss'],label="Training loss")
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```

```python
h.keys()
```

```python
plt.plot(h['val_accuracy'],label="validation acc")
plt.plot(h['accuracy'],label="Training acc")
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```

```python
y_=model.predict(x_test)
```

```python
y_.shape
```

```python
labels=[]
for i in range(y_.shape[0]):
    if (y_[i]>= 0.5):
        labels.append(1)
    else:
        labels.append(0)
```
